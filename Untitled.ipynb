```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas_datareader.data as web

from matplotlib.dates import DateFormatter
```

```python
def get_stock(stock_name, start_date, end_date, fix_name=None): 
    output = web.DataReader(f"{stock_name}", "yahoo", start_date, end_date)

    if fix_name == None:
        fix_name = stock_name.lower()
    
    print(f"{fix_name}:\t{str(output.index.min())[:10]} - {str(output.index.max())[:10]}")
    return output[['Close']].rename(columns={'Close': f"{fix_name}"}) 


def execute_merge(stocks):
    min_date_dict = dict()

    for stock in stocks:
        stock_name = stock.columns[0]
        min_date_dict[stock_name] = stock.index.min()

    key_list = list(min_date_dict.keys())
    val_list = list(min_date_dict.values())

    standard_stock_idx = val_list.index(min(min_date_dict.values()))
    output = stocks[standard_stock_idx]

    stocks.pop(standard_stock_idx)

    for stock in stocks:    
        output = output.merge(stock, how='left', left_index=True, right_index=True)
    
    return output 
```

```python
start_date = "2019-01-01"
end_date = "2021-03-15"

skyy = get_stock("SKYY", start_date, end_date)
wugi = get_stock("WUGI", start_date, end_date)
xle = get_stock("XLE", start_date, end_date)
edoc = get_stock("EDOC", start_date, end_date)
ufo = get_stock("UFO", start_date, end_date)
snp = get_stock("^GSPC", start_date, end_date, 'snp')
syld = get_stock("SYLD", start_date, end_date)
krw = get_stock("KRW=X", start_date, end_date, 'krw')
fvx = get_stock("^FVX", start_date, end_date, 'fvx')
tnx = get_stock("^TNX", start_date, end_date, 'tnx')
```

```python
pwr = get_stock("PWR", start_date, end_date)
```

```python
stocks = [skyy, wugi, xle, edoc, ufo, snp, syld, krw, fvx, tnx, pwr]

output = execute_merge(stocks)
output
```
